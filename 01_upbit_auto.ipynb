```python
# https://github.com/sharebook-kr/pyupbit
# https://blog.naver.com/zacra/223711423614
import pyupbit
import time
from dotenv import load_dotenv
import os

# 모든 암호화폐 목록
print(pyupbit.get_tickers())
```

```python
# 원화 거래만 하기 위해 원화 마켓 코인만 출력
Coins = pyupbit.get_tickers(fiat="KRW")
for coin in Coins:
    print(f"{coin} - {pyupbit.get_current_price(coin)}")
    time.sleep(0.1)
    # 매수/매도 호가
    # print(pyupbit.get_orderbook(coin))
    
```

# 시장가 매수

```python
# load .env
load_dotenv()

Access_Key = os.environ.get('Access_Key')
Secret_Key = os.environ.get('Secret_Key')
upbit = pyupbit.Upbit(Access_Key, Secret_Key)

for coin in Coins:
    #print(f"{coin} - {pyupbit.get_current_price(coin)}")
    time.sleep(0.1)

    if coin == "KRW-BTC" or coin == "KRW-TRX":
        upbit.buy_market_order(coin, 5000)
        print(f"{coin} - 구매완료")
```

# 시장가 매도


```python
# 입력받은 티커의 보유 수량 조회
btc_balance = upbit.get_balance("KRW-BTC")
print(btc_balance) 

# 비트코인을 시장가로 절반 판매(시장가 매도)
print(upbit.sell_market_order("KRW-BTC", btc_balance/2.0))
```

# 지정가 매수

```python
# 비트코인 현재 4800만원 x 10개 구매
# print(upbit.buy_limit_order("KRW-BTC", btc_now_price, 10))

# 비트코인 만원어치 구매 10000(보유금액)/48000000
# print(upbit.buy_limit_order("KRW-BTC", btc_now_price, won/btc_now_price))

# https://docs.upbit.com/docs/krw-market-info
# btc_now_price가 소수점이면 ex) 99.998을 넣으면 동작 안할 수도 있다
# 거래소에서 소수점 몇자리까지 허용하는지 확인해야함
# 업비트 -  호가가 1,000원 일 경우 999.1원, 999.2원, 999.3원, ... , 1,001원, 1,002원, 1,003원 으로 주문을 넣을 수 있으며, 1,000.1원, 1,000.2원 등의 가격으로는 주문할 수 없습니다.


load_dotenv()

Access_Key = os.environ.get('Access_Key')
Secret_Key = os.environ.get('Secret_Key')
upbit = pyupbit.Upbit(Access_Key, Secret_Key)

btc_now_price = pyupbit.get_current_price("KRW-BTC")    # 현재 비트코인 가격
btc_target_price = btc_now_price * 0.998                   # 0.02퍼신트 떨어진 가격으로
won = 10000                                             # 원금 10000원워치 구매

# 10000원어치 비트코인 지정가 매수
print(upbit.buy_limit_order("KRW-BTC", pyupbit.get_tick_size(btc_target_price), won/btc_target_price))
```
