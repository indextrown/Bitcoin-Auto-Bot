```python
# https://blog.naver.com/zacra/223404688692 
import pyupbit
from dotenv import load_dotenv
import os
import json

# load .env
load_dotenv()

# key
Access_Key = os.environ.get('Access_Key')
Secret_Key = os.environ.get('Secret_Key')
upbit = pyupbit.Upbit(Access_Key, Secret_Key)
```

```python
my_balances = upbit.get_balances()
pretty_balances = json.dumps(my_balances, indent=4, ensure_ascii=False)
print(pretty_balances)
```

```python
for coin_balance in my_balances:

    try:
        # 티커 이름
        ticker = coin_balance["currency"]

        # 기저사례
        if ticker == "KRW" or ticker == "VTHO" or ticker == "APENFT":
            continue

        # 현재 마켓 코인 가격
        nowPrice = pyupbit.get_current_price("KRW-"+ticker)

        # 평균 매입 단가
        avg_price = float(coin_balance['avg_buy_price'])
        
        # 수익률
        revenu_rate = (nowPrice - avg_price) / avg_price * 100.0

        print(coin_balance['currency']) 
        print(f"잔고: {coin_balance['balance']}")
        print(f"평균 매입 단가: {avg_price}")
        print(f"수익률: {revenu_rate}")
        print()

        # 수익률이 1.7보다 크면
        if revenu_rate >= 1.7:
            print("!!!!!!!!!수익발생!!!!!!!!!!")
            #  현재가격으로 팔겠다
            # upbit.sell_limit_order("KRW-", ticker, nowPrice)

            # 현재가격보다 0.1퍼센트 위의 금액으로 팔고싶다
            # print(upbit.sell_limit_order("KRW-", ticker, pyupbit.get_tick_size(nowPrice * 1.002), coin_balance['balance']))

    except Exception as e: 
        print("error: ", e)
```
