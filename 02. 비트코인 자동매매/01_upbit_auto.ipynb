```python
# https://github.com/sharebook-kr/pyupbit
# https://blog.naver.com/zacra/223711423614

import pyupbit
import time
import os
from dotenv import load_dotenv

```

## 모든 암호화폐 목록

```python
# 모든 암호화폐 목록
# print(pyupbit.get_tickers())

# 원화 거래만 하기 위해 원화 마켓 코인만 출력
coins = pyupbit.get_tickers(fiat="KRW")
for coin in coins:
    print(f"{coin} - {pyupbit.get_current_price(coin)}")
    time.sleep(0.05) # 초당 10회: Upbit Quotation API 권장치
```

## 시장가 매수

```python
# .env 파일에서 환경변수 로드
load_dotenv()
Access_Key = os.environ.get('ACCESS_KEY')
Secret_Key = os.environ.get('SECRET_KEY')

print(f"Access Key: {Access_Key}")
print(f"Secret Key: {Secret_Key}")

upbit = pyupbit.Upbit(Access_Key, Secret_Key)

for coin in coins:
    #print(f"{coin} - {pyupbit.get_current_price(coin)}")
    time.sleep(0.1)

    if coin == "KRW-BTC" or coin == "KRW-TRX":
        upbit.buy_market_order(coin, 5000)
        print(f"{coin} - 구매완료")
```
