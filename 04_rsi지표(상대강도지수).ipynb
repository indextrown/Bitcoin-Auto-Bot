## 일정 기간 전일 대비
#### RSI지표 =  상승분의 평균값 / (상승분의 평균값 + 하락분의 평균값)


# RSI 지표 (Relative Strength Index)

RSI는 특정 기간 동안의 **가격 상승**과 **하락**을 비교하여 **과매수 (Overbought)** 또는 **과매도 (Oversold)** 상태를 판단하는 데 도움을 주는 기술적 분석 지표입니다.

---

## RSI 공식

1. **RS (Relative Strength)** 계산:
   ```
   RS = 상승분의 평균값 / 하락분의 평균값
   ```

2. **RSI 계산**:
   ```
   RSI = 100 - (100 / (1 + RS))
   ```

---

## 계산 단계

1. **상승분과 하락분 계산**:
   - 특정 기간 동안 (일반적으로 14일)의 가격 변화에서 상승분과 하락분을 분리합니다.
   - 상승분은 가격이 상승한 날의 값만 사용하고, 하락분은 가격이 하락한 날의 값만 사용합니다.

2. **평균값 계산**:
   - **상승분의 평균값**: 최근 14일 동안 상승한 가격의 평균.
   - **하락분의 평균값**: 최근 14일 동안 하락한 가격의 평균.

3. **RS 계산**:
   ```
   RS = 상승분의 평균값 / 하락분의 평균값
   ```

4. **RSI 계산**:
   ```
   RSI = 100 - (100 / (1 + RS))
   ```

---

## RSI 해석

- **RSI > 70**: 과매수 상태 (가격 하락 가능성 증가)
- **RSI < 30**: 과매도 상태 (가격 상승 가능성 증가)

---

## 예제

1. **14일 동안의 데이터**:
   - 상승분의 총합: 140
   - 하락분의 총합: 60

2. **평균 상승분**:
   ```
   평균 상승분 = 140 / 14 = 10
   ```

3. **평균 하락분**:
   ```
   평균 하락분 = 60 / 14 ≈ 4.29
   ```

4. **RS 계산**:
   ```
   RS = 10 / 4.29 ≈ 2.33
   ```

5. **RSI 계산**:
   ```
   RSI = 100 - (100 / (1 + 2.33)) ≈ 70.01
   ```

---

RSI는 **추세와 모멘텀**을 함께 파악하는 데 유용하며, 다른 기술적 지표와 함께 사용하면 더욱 효과적입니다.


````python
import pyupbit
from dotenv import load_dotenv
import os
import pandas as pd

# load .env
load_dotenv()

# key
Access_Key = os.environ.get('Access_Key')
Secret_Key = os.environ.get('Secret_Key')
upbit = pyupbit.Upbit(Access_Key, Secret_Key)

#RSI지표 수치를 구해준다. 첫번째: 분봉/일봉 정보, 두번째: 기간
# 지정한 기간 동안의 전일 대비 상승분의 평균 / (전일대비 상승분의 평균+하락분의 평균)
def GetRSI(ohlcv,period):
    ohlcv["close"] = ohlcv["close"]
    delta = ohlcv["close"].diff()
    up, down = delta.copy(), delta.copy()
    up[up < 0] = 0
    down[down > 0] = 0
    _gain = up.ewm(com=(period - 1), min_periods=period).mean()
    _loss = down.abs().ewm(com=(period - 1), min_periods=period).mean()
    RS = _gain / _loss
    return pd.Series(100 - (100 / (1 + RS)), name="RSI")

#비트코인의 일봉 정보를 가져온다.  
df = pyupbit.get_ohlcv("KRW-BTC",interval="day")

# RSI지표를 계산해서 출력한다.
print(GetRSI(df, 14))

# 어제 RSI14 지표
rsi14 = float(GetRSI(df,14).iloc[-2])
print("\n어제 RSI:", rsi14)

# 오늘 RSI14 지표
rsi14 = float(GetRSI(df,14).iloc[-1])
print("실시간 RSI:", rsi14)



# RSI14에서 오늘 지표는 계속 변하기 떄문에 오늘 지표를 활용할건지 안할건지 개발자가 판단
````
